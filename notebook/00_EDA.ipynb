```python
import pandas as pd
```

```python
df_metrics = pd.read_excel(
    r"../data/rappiData.xlsx",
    sheet_name='RAW_INPUT_METRICS'
)

df_orders = pd.read_excel(
    r"../data/rappiData.xlsx",
    sheet_name='RAW_ORDERS'
)
```

```python
nulls_per_row = df_orders.isna().sum(axis=1)
df_orders.loc[nulls_per_row.sort_values(ascending=False).head(5).index]
```

```python
nulls_per_row = df_orders.isna().sum(axis=1)

count_by_nulls = nulls_per_row.value_counts().sort_index()

count_1_to_9 = count_by_nulls.loc[1:9]
count_1_to_9
```

```python
nulls_per_row = df_metrics.isna().sum(axis=1)
df_metrics.loc[nulls_per_row.sort_values(ascending=False).head(5).index]
```

```python
nulls_per_row = df_metrics.isna().sum(axis=1)

count_by_nulls = nulls_per_row.value_counts().sort_index()

count_1_to_9 = count_by_nulls.loc[1:9]
count_1_to_9
```
